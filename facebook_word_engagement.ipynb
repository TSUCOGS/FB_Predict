```python
from setup import *
import glob
import os
import nltk
import datetime
import pickle
import joblib
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from keras.preprocessing.text import hashing_trick
from keras.preprocessing.text import text_to_word_sequence
import string
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
import sys
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.embeddings import Embedding
from keras.callbacks import EarlyStopping
```

```python
file_list = glob.glob("Facebook Insights*")
print(file_list)
```

```python
df_list = []

for file in file_list:
    df_list.append(pd.read_csv(file,encoding='iso-8859-1').loc[1:])
    
# for df in df_list:
#     display(df)
    
df = pd.concat(df_list).reset_index()
df.to_csv('All_Files.csv')
```

```python
df.columns
```

```python
short_df = df[['Post ID','Post Message','Lifetime Post Total Reach','Lifetime Engaged Users']].dropna()
tokenized_df = short_df
```

```python
print(stopwords.words('english'))
```

```python
tokenized_df['Sentences'] = tokenized_df['Post Message'].transform(lambda x: sent_tokenize(str(x)))
tokenized_df['Words'] = tokenized_df['Post Message'].transform(lambda x: ' '.join([WordNetLemmatizer().lemmatize(y.lower()) for y in word_tokenize(x) if y.isalpha()]))
tokenized_df['Words'] = tokenized_df['Words'].transform(lambda x: ' '.join([y for y in word_tokenize(x) if y not in stopwords.words('english')]))
```

```python
display(tokenized_df)
```

```python
#remove entries with lower than 50 reach, likely to skew the proportions
good_proportion = tokenized_df
print(good_proportion.columns)
good_proportion['Prop'] = pd.to_numeric(good_proportion['Lifetime Engaged Users'])/pd.to_numeric(good_proportion['Lifetime Post Total Reach'])
good_proportion = tokenized_df.loc[pd.to_numeric(good_proportion['Lifetime Post Total Reach']) >= 50] 
```

```python
display(good_proportion)
```

```python
display(good_proportion.sort_values('Prop', ascending = False))
```

```python
df['Time'] = pd.to_datetime(df['Posted'])
df['Time'] = df['Time'].apply(lambda x: (x.hour+2)%24)
```

```python
plt.hist(df['Time'], bins=24)
plt.show()
```

```python
X_all = good_proportion['Words']
y_all = good_proportion['Lifetime Engaged Users']
```

```python
good_proportion['Words'].apply(lambda l: len(set(text_to_word_sequence(str(l),filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',lower=True,split=" ")))).sort_values(ascending = False) # this is the post with the largest number of unique words not in the stop_
```

```python
Tfidf_vectorizer = TfidfVectorizer(stop_words = 'english', min_df = 5, max_df = 0.25, ngram_range = (1,2)) #words must be in at least 5 posts and no more than 25% of all posts
Tfidf_vectorizer.fit(X_all)

joblib.dump(Tfidf_vectorizer, 'vectorizer.pkl')

transformed_X = Tfidf_vectorizer.transform(X_all)
```

```python
transformed_X.shape
```

# Keras Stuff

```python
all_transformed_X = transformed_X
all_y = y_all

X_train, X_test, y_train, y_test = train_test_split(all_transformed_X, all_y, test_size=0.25, random_state=42)


n_cols = all_transformed_X.shape[1]
model = Sequential()
model.add(Dense(250, activation='relu', input_shape = (n_cols,)))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer='adam', loss='mean_squared_error')


# early_stopping_monitor = EarlyStopping(patience=5)
model1 = model.fit(all_transformed_X, all_y, epochs=100, validation_split=0.3, verbose = False)
```

```python
plt.plot(model1.history['val_loss'], 'r')
plt.xlabel('Epochs')
plt.ylabel('Validation score')
plt.show()
```

```python
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model.h5")
print("Saved model to disk")
```
